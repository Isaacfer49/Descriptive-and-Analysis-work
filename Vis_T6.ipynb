# Renta petrolera en México.
En esta parte, haremos el analisis correspondiente de la renta petrolera México.


```python
filter = data1['Country Name'] == 'Mexico'
Mexico= data1.where(filter).dropna()
Mexico
```

```python
plt.rcParams["figure.figsize"] = (10,10)
x = Mexico.transpose()
x.columns = x.iloc[0]
x.drop('Country Name').plot()
plt.show()
```

```python
Mexico = Mexico.transpose()
Mexico.columns = Mexico.iloc[0]
```

```python
Mexico.drop(index=Mexico.index[0], axis=0, inplace=True)
```

```python
x = Mexico.reset_index()[0].values 
```

```python
y = Mexico.reset_index()['Mexico'].values
```

```python
Mex = pd.DataFrame()
Mex['Anos'] = x
Mex['GDP'] = y
```

```python
P1 = (Mex['Anos'] >= 1970) & (Mex['Anos']  <= 1976)
```

```python
seventys = Mex.where(filter).dropna()
P2 = (Mex['Anos'] >= 1976) & (Mex['Anos'] <= 1982)
P3 = (Mex['Anos'] >= 1982) & (Mex['Anos'] <= 1988)
P4 = (Mex['Anos'] >= 1988) & (Mex['Anos'] <= 1994)
P5 = (Mex['Anos'] >= 1994) & (Mex['Anos'] <= 2000)
P6 = (Mex['Anos'] >= 2000) & (Mex['Anos'] <= 2006)
P7 = (Mex['Anos'] >= 2006) & (Mex['Anos'] <= 2012)
P8 = (Mex['Anos'] >= 2012) & (Mex['Anos'] <= 2018)
P9 = (Mex['Anos'] >= 2018) & (Mex['Anos'] <= 2020)
```

```python
periodos = [P1,P2,P3,P4,P5,P6,P7,P8,P9]
x = []
for p in periodos:
  val= Mex.where(p).dropna()['GDP'].sum()
  x.append(val)
```

```python
x = pd.DataFrame(x)
x['Periodo'] = ['P1','P2','P3','P4','P5','P6','P7','P8','P9']
x.sort_values(by = 0,ascending=False)
```
