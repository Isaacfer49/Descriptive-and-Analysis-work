
## Contextualización.
En Estados Unidos, el SAT (Scholastic Assessment Test) es un examen estandarizado que se utiliza comúnmente como parte del proceso de admisión a la universidad. El examen se enfoca en evaluar las habilidades académicas de los estudiantes en materias como matemáticas, lectura y escritura.

El SAT es administrado por el College Board, una organización sin fines de lucro que también administra otros exámenes estandarizados, como el PSAT/NMSQT y el Advanced Placement (AP). El examen SAT se ofrece varias veces al año en centros de pruebas autorizados en todo el país.

El SAT es utilizado por muchas universidades como un factor para evaluar la admisión de estudiantes potenciales, junto con otros criterios como calificaciones, ensayos y actividades extracurriculares. Las puntuaciones del SAT también se pueden usar para solicitar becas y otros programas de ayuda financiera.

En resumen, el SAT es un examen estandarizado que se utiliza como parte del proceso de admisión a la universidad en Estados Unidos. Sirve como una medida de las habilidades académicas de los estudiantes y es utilizado por muchas universidades como un factor en la toma de decisiones de admisión.

```python
import plotly.express as px
import pandas as pd

# Load data
states = pd.read_csv('/content/States.csv')

states.head(5)

```

```python
states = states.rename(columns={"Unnamed: 0":"States"})
```

```python
import plotly.express as px
import pandas as pd




# Create choropleth map
fig = px.choropleth(states, locations="States", color="dollars",
                    hover_name="States",locationmode="USA-states",scope="usa", 
                    color_continuous_scale=px.colors.sequential.Plasma)

# Display map
fig.show()
```

```python

# Create choropleth map
fig = px.choropleth(states, locations="States", color="SATV",
                    hover_name="States",locationmode="USA-states",scope="usa", 
                    color_continuous_scale=px.colors.sequential.Plasma)

# Display map
fig.show()
```

```python

# Create choropleth map
fig = px.choropleth(states, locations="States", color="SATM",
                    hover_name="States",locationmode="USA-states",scope="usa", 
                    color_continuous_scale=px.colors.sequential.Plasma)

# Display map
fig.show()
```

```python
states['text']='SATv '+ states['SATV'].astype(str)+'SATm '+ states['SATM'].astype(str) +'<br>'+'State '+ states['States']
```

```python
states['full_SAT_score'] = states['SATM'] + states['SATV']
```

```python
# Create choropleth map
fig = px.choropleth(states, locations="States", color="full_SAT_score",
                    hover_name="States",locationmode="USA-states",scope="usa", 
                    color_continuous_scale=px.colors.sequential.Plasma)

# Display map
fig.show()
```

* Massachusetts - Puntuación promedio: 1124

Es importante tener en cuenta que estas puntuaciones son promedios estatales y que las puntuaciones individuales pueden variar según la escuela y el estudiante.

Sin embargo, es importante señalar que las puntuaciones del SAT no son la única medida del éxito académico y que otros factores también influyen en la admisión a la universidad. Además, es posible que existan diferencias socioeconómicas, culturales y educativas entre los estados que puedan influir en las puntuaciones del SAT.

Por lo tanto, aunque es útil conocer las puntuaciones promedio del SAT en diferentes estados, no es una medida definitiva del éxito académico y la capacidad de un estudiante para tener éxito en la universidad o en la vida en general.


# Snow inventory

```python
snow = pd.read_csv('/content/snow_inventory.csv')
```

```python
snow.head(5)
```

```python
snow = snow.rename(columns={"-14.3167":"Longitude","-170.7667":"Latitude"})
```

```python
snow = snow.rename(columns={"408.4":"other"})
```

```python
import plotly.graph_objects as go



# Create graph
fig = go.Figure(data=go.Scattergeo(
    lat = snow['Longitude'],
    lon = snow['Latitude'],
    text=snow['other'],
    mode = 'markers',
    marker = dict(
        size = 10,
        color = snow['other'],
        colorscale='Viridis',
        opacity = 0.8,
        colorbar=dict(title='Elevación'),
        line = dict(
            width = 1,
            color = 'rgba(0, 0, 0, 0.8)'
        )
    )
))

# Customize layout
fig.update_layout(
    title = 'Elevación en los Estados Unidos',
    geo = dict(
        scope = 'usa',
        projection_type = 'albers usa',
        showland = True,
        landcolor = 'rgb(217, 217, 217)',
        subunitwidth=1,
        countrycolor='rgb(255, 255, 255)'
      
)
)
```

La latitud de los Estados Unidos es muy diversa debido a la gran extensión territorial del país que se extiende desde la latitud 71.3°N en Alaska hasta la latitud 18.9°N en Hawái. Por lo tanto, la latitud en los Estados Unidos tiene una gran influencia en el clima, la geografía y la biodiversidad del país.

En general, los estados ubicados en latitudes más bajas (como Florida, Texas, California y Hawái) tienen un clima más cálido y húmedo, mientras que los estados ubicados en latitudes más altas (como Alaska, Montana, Dakota del Norte y Maine) tienen un clima más frío y seco.

Además, la latitud también influye en la duración de los días y las noches en diferentes épocas del año. En el norte del país, como Alaska, los días son más largos en el verano y más cortos en el invierno, mientras que en el sur, como en Florida, los días y las noches tienen una duración más uniforme durante todo el año.

En conclusión, la latitud tiene una gran influencia en el clima, la geografía y la biodiversidad de los Estados Unidos, lo que a su vez afecta la vida de las personas y las culturas regionales en el país.
